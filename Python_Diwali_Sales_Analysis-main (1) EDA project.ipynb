```python
# import python libraries

import numpy as np 
import pandas as pd 

## visualizing libraries

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# import csv file
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
# drop blank columns:
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
df.info()
```

```python
#check null values:
pd.isnull(df).sum()
```

```python
#### there are 12 null values in amount column so we have to remove all null values
```

```python
df.dropna(inplace=True) #inplace is used for save the code 
```

```python
df.shape
```

```python
# change the data type because after drop null values we have only one value imn float
df['Amount'] = df['Amount'].astype('int')
```

```python
df.columns
```

```python
## rename columns:
df.rename(columns= {'Marital_Status': 'Shaadi'})
```

```python
## mathematical discription like MEAN,MEDIAN.MODE.....
df.describe()
```

# Exploratory Data Analysis


GENDER

```python
df.columns
```

```python

ax = sns.countplot(x= 'Gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
# plotting a bar chart for gender vs total amount

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)
```

from this graph we observe that most buyers are female and purchasing power is also more than men

# STATE

```python
df.columns
```

```python
# Total number of orders from top 10 states:
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
```

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
```

```python
# total amount/sales from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount')
```

 from this graph we observe that most orders and totalsales/amount are from UP,MH,KR...

# MARITAL STATUS

```python
ax = sns.countplot(data = df, x = 'Marital_Status')
```

```python
ax = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(10,8)})
for bars in ax.containers:
    ax.bar_label(bars)
```

# OCCUPATION

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation')
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```
